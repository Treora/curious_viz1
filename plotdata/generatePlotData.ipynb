```python
# 1-dimensional case
def ptilde_1d(xtilde, x_data, sigma_n):
    return np.mean(gaussian.pdf(xtilde-x_data,0,sigma_n))
    
def denoising_g1d(xtilde, x_data, sigma_n):
    num = np.mean(x_data*gaussian.pdf(xtilde-x_data,0,sigma_n))
    den = np.mean(gaussian.pdf(xtilde-x_data,0,sigma_n))
    return num/den
```

```python
from scipy.stats import multivariate_normal as gaussian_multi
# PDF for mixture of Gaussians. ws =[w1,w2,...], mus = [mu1, mu2, ...] etc.
def MixtGaussian_dist(x, ws, mus, sigmas):
    d = mus[...,None].shape[1]
    list_gauss = [w*gaussian_multi.pdf(x, mu, sigma*np.identity(d))
                  for w, mu, sigma in zip(ws, mus, sigmas)]
    return sum(list_gauss)

# Optimal denoising function for mixture of Gaussians.
# ws =[w1,w2,...], mus = [mu1, mu2, ...] etc.
# USEFUL IN ALL DIMENSIONS
def denoising_g_mixt(x, ws, mus, sigmas, sigma_ns):
    d = mus[...,None].shape[1]
    sigma_tots = np.sqrt(sigmas**2+sigma_ns**2)
    vs = sigmas**2 / sigma_tots**2
    C0s = np.array([w*gaussian_multi.pdf(x, mu, (sigma_tot**2)*np.identity(d))
          for w, mu, sigma_tot in zip(ws, mus, sigma_tots)],dtype='float128')
    #print(C0s.dtype)
    if d >1:    
        C1s = np.array([(v*x + (1-v)*np.broadcast_to(mu, x.shape))*C0[...,None]
                   for v, mu, C0 in zip(vs, mus, C0s)],dtype='float128')
    else:
        C1s = np.array([(v*x + (1-v)*np.broadcast_to(mu, x.shape))*C0
                   for v, mu, C0 in zip(vs, mus, C0s)],dtype='float128')   
    numerator = sum(C1s)
    denominator = sum(C0s)   
    if d > 1:        
        return np.array(numerator / denominator[..., None],dtype='float64')
    else:
        return np.array(numerator / denominator,dtype='float64')
    
# BELOW VERSIONS SHOULD BE VECTORIZED W.R.T X
# ONLY USEFUL IN 1D
def denoising_g_mixt_vect(x, ws, mus, sigmas, sigma_ns):
    d = mus[...,None].shape[1]
    sigma_tots = np.sqrt(sigmas**2+sigma_ns**2)
    vs = sigmas**2 / sigma_tots**2
    C0s = np.array([gaussian_multi.pdf(x, mu, (sigma_tot**2)*np.identity(d))
          for mu, sigma_tot in zip(mus, sigma_tots)],dtype='float128')
    C1s = np.array([(v*x + (1-v)*mu)*C0
                   for v, mu, C0 in zip(vs, mus, C0s)],dtype='float128')
    numerator = np.sum(ws*C1s)
    denominator = np.sum(ws*C0s)
    return np.array(numerator / denominator,dtype='float64')
   
def denoising_g_mixt_vect2(x, w, mu, sigma, sigma_n):
    sigma_tot = np.sqrt(sigma**2+sigma_n**2)
    v = sigma**2 / sigma_tot**2
    C0 = gaussian.pdf(x, mu, sigma_tot)
    C1 = (v*x + (1-v)*mu)*C0
    numerator = np.sum(w*C1)
    denominator = np.sum(w*C0)
    return numerator / denominator

# analytic denoising function for mixture of two gaussians
def denoising_g__mixt_vect_WRONG(x, w, mu, sigma, sigma_n):
    sigma_tot = np.sqrt(sigma**2+sigma_n**2)
    v = sigma**2 / sigma_tot**2
    C0 = sigma*gaussian.pdf(x, mu, sigma_tot)
    C1 = (1-v)*(x-mu)*C0
    numerator = np.sum(w*C1)
    denominator = np.sum(w*C0)
    return x - numerator / denominator
```

```python
# generate data for uniform distribution
nsamples = 10000

# varied parameter values
sigmas = [0.5, 1.0, 1.5, 2.0, 2.5]
sigma_n = 1.0
mu = 0

d_uniform = []

for j,sigma in enumerate(sigmas):
    # derived parameters (start and end of uniform density)
    x2 = mu + np.sqrt(3)*sigma
    x1 = mu - np.sqrt(3)*sigma
    # corrupted and uncorrupted data distributions
    p_dist = lambda x: np.piecewise(x, [(x < x1) | (x > x2),
                                     (x >= x1) & (x <= x2)],[0,1/(x2-x1)])
    ptilde_dist = lambda xtilde: (gaussian.cdf(x2, xtilde, sigma_n)\
                                  -gaussian.cdf(x1, xtilde, sigma_n))/(x2-x1)
    # generate samples
    x_data = x1 + (x2-x1)*np.random.rand(nsamples)
    xtilde_data = x_data + np.random.randn(nsamples)*sigma_n
    # optimal denoising function
    denoiser = lambda xtilde_: denoising_g1d(xtilde_, x_data, sigma_n)
    g = np.vectorize(denoiser)
    d_uniform.append({
        'xDomain': [xmin, xmax],
        'sigma': sigma,
        'data': list(p_dist(x_points)),
        'noisy': list(ptilde_dist(x_points)),
        'denoise': list(g(x_points)),
    })

toJavascript('uniform', d_uniform)
```
